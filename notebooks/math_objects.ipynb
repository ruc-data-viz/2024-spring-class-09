# Math Objects - Complex Data

Complex functions are canonically visualized as surfaces. Here we will see some examples of how to visualize these with matplotlib.

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 16
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
plt.rcParams["text.usetex"] = True

def plot_complex_function(complex_function, limit=5):
    """Plots a complex function, accepting a callable as input"""

    # Generate complex numbers for the grid
    x = np.linspace(-limit, limit, 500)
    y = np.linspace(-limit, limit, 500)
    x_grid, y_grid = np.meshgrid(x, y)
    z_grid = x_grid + 1j * y_grid

    complex_function_values = complex_function(z_grid)

    # Plotting real part
    plt.figure(figsize=(12, 8))

    plt.subplot(1, 2, 1)
    plt.imshow(
        np.real(complex_function_values), extent=[-limit, limit, -limit, limit], cmap="nipy_spectral", origin="lower"
    )
    plt.colorbar()
    plt.title(r"$\mathrm{Re}\{f(z)\}$")
    plt.xlabel(r"$x$")
    plt.ylabel(r"$iy$")
    plt.grid(True, linestyle="--", alpha=0.7)

    # Plotting imaginary part
    plt.subplot(1, 2, 2)
    plt.imshow(
        np.imag(complex_function_values), extent=[-limit, limit, -limit, limit], cmap="nipy_spectral", origin="lower"
    )
    plt.colorbar()
    plt.title(r"$\mathrm{Im}\{f(z)\}$")
    plt.xlabel(r"$x$")
    plt.ylabel(r"$iy$")
    plt.grid(True, linestyle="--", alpha=0.7)

    plt.show()
```

## Sine

```python
plot_complex_function(np.sin, limit=4)
```

## Cosine

```python
plot_complex_function(np.cos, limit=4)
```

## Tangent

```python
plot_complex_function(np.exp, limit=4)
```

```python
def complex_function(x):
    return np.log(1 / x)


plot_complex_function(complex_function, limit=4)
```

```python
def complex_function(x):
    return np.sqrt(x)


plot_complex_function(complex_function, limit=4)
```

```python
def complex_function(x):
    return np.log(1 / (x**x))


plot_complex_function(complex_function, limit=4)
```
